## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________
Inner Join - keeps only the overlaps
Full Outer Join - keeps everything
Left (Outer) Join - keeps everything from the first table and overlaps but drops items unique to the second table.
Right (Outer) Join - keeps everything from the second table and overlaps but drops items unique to the first table.
Left Antijoin - keeps everything from the first table except ones that matches the second table
Right Antijoin - keeps everything from the second table except ones that matches the first table 



## Question 5

Explain the relationship between Aggregate functions and GROUP BY statements.

Type your answer below this line:
_______________________________________________________________________________________________________________________________
Aggregate function 'aggregates' data to perform a calculation. GROUP BY statement provides information to base such aggregation on. If we want the sum of individual groups, calculating the sum of individual groups is the role of aggregate, while deciding how to divide each group is the role of GROUP BY.
